```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df=pd.read_csv("House Price India.csv")
```

```python
df.head()
```

```python
#Corelation 
df.corr()
```

```python
# Scatter plot for the Condition of house and  the price
x=df['condition of the house']
y=df['Price']
plt.scatter(x,y)
plt.xlabel('Condition of the house')
plt.ylabel('Price ')
plt.title('Scatter plot of Condition of th house and Price')
plt.show()
```

```python
# Scatter plot for the Condition of house and  the price
x=df['number of bedrooms']
y=df['number of bathrooms']
plt.scatter(x,y)
plt.xlabel('Number of bedrooms')
plt.ylabel('Number of bathrooms ')
plt.title('Scatter plot of Number of bedrooms and number of bathrooms')
plt.show()
```

```python
# Scatter plot for LLiving area and lot area
x=df['living area']
y=df['lot area']
plt.scatter(x,y)
plt.xlabel('Living area')
plt.ylabel('lot area')
plt.title('Scatter plot for Living area and lot area')
plt.show()
```

```python
 #Bar plot for LLiving area and lot area
x=df['living area']
y=df['lot area']
plt.bar(x,y)
plt.xlabel('Living area')
plt.ylabel('lot area')
plt.title('bar for Living area and lot area')
plt.show()
```

```python
# Bar for the Condition of house and  the price
x=df['number of bedrooms']
y=df['number of bathrooms']
plt.bar(x,y)
plt.xlabel('Number of bedrooms')
plt.ylabel('Number of bathrooms ')
plt.title('Bar of Number of bedrooms and number of bathrooms')
plt.show()
```

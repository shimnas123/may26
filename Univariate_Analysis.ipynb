```python
con_of_house=df[df['condition of the house']=='Postal Code']
#create a frequency table
price =con_of_house['Price'].value_counts().sort_index()
price
```

```python
#plot the frequency distribution as a histogram
plt.hist(con_of_house['Price'], bins=20,edgecolor='blue')
plt.title("Condition of house According to postal area Price Distribution")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
# plot histogram off Bathroom variable
plt.hist(df['number of bathrooms'], bins=20, density=True, alpha=0.6, color='g')
plt.title('No of bathroom distribution')
plt.xlabel('Bathroom')
plt.ylabel('Frequency')
plt.show()
```

```python
# plot histogram ffor number of  room distribution
plt.hist(df['number of bedrooms'], bins=20, density=True, alpha=0.6, color='b')
plt.title('NO. of bedrooms Distribution')
plt.xlabel('BedRooms')
plt.ylabel('Frequency')
plt.show()
```

```python
#histogram for each numerical variable
df.hist(bins=10, figsize=(10,10))
plt.show()
```

```python
#boxplot for each numerical value
df.plot(kind='box', subplots=True,figsize=(15,8))
plt.show()
```

```python

# create a boxplot of the scores variable
plt.boxplot(df['Price'])
plt.title('Boxplot of the house Price')
plt.show()

# calculate the quartiles and interquartile range
q1, q3 = df['Price'].quantile([0.25, 0.75])
iqr = q3 - q1
# identify outliers based on the boxplot rule
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]

# print the outliers
print('Outliers:')
print(outliers)

```
